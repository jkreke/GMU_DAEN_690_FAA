```python
import pandas as pd
df = pd.read_csv('/Users/Char/Desktop/laser_combined_05Apr.csv')
df.head(n=5)
```

```python
import re
```

```python
#create new column with distance from landmark (airport or other) reported in remarks as NM

NM = ['NM']
df['Dist'] = df['remarks'].str.findall(r'[\d]+ (?i)\b(?:{})\b'.format('|'.join(NM))).apply(', '.join)
```

```python
#check to make sure new column added
df.head()
```

```python
#create separate columns to pull out #different ways they reported distance (note some reported as NM were not
# included in original "dist" column above, tried again, varying the regex pattern slighltly)
#these were later examined in excel

mile = ['mile']
df['mile'] = df['remarks'].str.findall(r'[\d]+ (?i)\b(?:{})\b'.format('|'.join(mile))).apply(', '.join)

miles = ['miles']
df['miles'] = df['remarks'].str.findall(r'[\d]+ (?i)\b(?:{})\b'.format('|'.join(miles))).apply(', '.join)

NM2 = ['NM']
df['NM2'] = df['remarks'].str.findall(r'[\d]+(?i)\b(?:{})\b'.format('|'.join(NM2))).apply(', '.join)
```

```python
#check to make sure new columns added
df.head()
```

```python
#write output to file

df.to_csv('/Users/Char/Desktop/lasers2.csv', index = False)
```
