```python
df = pd.read_csv(" ")
```

```python
df_mapping = pd.DataFrame({
    'skyc1': ['CLR', 'FEW', 'SCT', 'BKN', 'OVC', 'VV ', 'M'],
})
sort_mapping = df_mapping.reset_index().set_index('skyc1')
df['skyc1_num'] = df['skyc1'].map(sort_mapping['index'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df, x='skyc1', hue='HotSpot', multiple="stack")
plt.xlabel('Cloud Coverage')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Cloud Coverage for Strikes')
```

```python
dfTMPF = df
```

```python
indexNames = dfTMPF[dfTMPF['tmpf'] == 'M'].index
dfTMPF.drop(indexNames , inplace=True)
```

```python
dfTMPF['tmpf']=dfTMPF['tmpf'].astype(str).astype(float)
```

```python
sns.histplot(data=dfTMPF, x='tmpf', bins=50, hue='HotSpot', multiple="stack")
plt.xlabel('Temperature in Fahrenheit')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Temperature for Strikes')
```

```python
dfVSB = df
```

```python
indexNames = dfVSB[dfTMPF['vsby'] == 'M'].index
dfVSB.drop(indexNames , inplace=True)
```

```python
dfVSB['vsby']=dfTMPF['vsby'].astype(str).astype(float)
```

```python
sns.histplot(data=dfVSB, x='vsby', bins=10, hue='HotSpot', multiple="stack")
plt.xlabel('Visibility in Statute Miles, max 10 SM')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Visibility for Strikes')
```

```python
dfRELH = df
```

```python
indexNames = dfRELH[dfRELH['relh'] == 'M'].index
dfRELH.drop(indexNames , inplace=True)
```

```python
dfRELH['relh']=dfRELH['relh'].astype(str).astype(float)
```

```python
sns.histplot(data=dfRELH, x='relh', bins=50, hue='HotSpot', multiple="stack", legend= False)
plt.xlabel('Relative Humidity')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Relative Humidity for Strikes')
```

```python
sns.histplot(data=dfN, x='week', hue='HotSpot', multiple="stack", bins=53, legend =False)
plt.xlabel('Week Number')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Strikes by Week')
```

```python
sns.histplot(data=dfN, x='hour', hue='HotSpot', multiple="stack", legend=False, bins=24)
plt.xlabel('Hour Number')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Strikes by Local Hour')
```

```python
sns.histplot(data=dfN, x='month', hue='HotSpot', multiple="stack", legend=False, bins=12)
plt.xlabel('Month Number')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Strikes by Month')
```

```python
sns.histplot(data=dfN, x='wday', hue='HotSpot', multiple="stack", bins=7)
plt.xlabel('Day of Week; 1-Sun - 7-Sat')
plt.ylabel('Count by Laser Strike Incident')
plt.title('Strikes by Week Day')
```
